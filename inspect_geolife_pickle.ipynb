# Inspect Geolife Cluster Pickle

Load the `RLSTCcode/data/geolife_clustercenter` pickle and inspect its structure for plotting.

```python
import pickle
from pathlib import Path

path = Path('RLSTCcode/data/geolife_clustercenter')
print('Exists:', path.exists())
with path.open('rb') as f:
    data = pickle.load(f)

print('Type:', type(data))
print('Len:', len(data) if hasattr(data, '__len__') else 'n/a')
print('Sample repr:', repr(data)[:500])

if isinstance(data, list) and len(data) > 0:
    first = data[0]
    print('first type:', type(first))
    try:
        print('first len:', len(first))
    except Exception as e:
        print('first len error', e)

    if isinstance(first, (list, tuple)) and len(first) > 2:
        cluster_dict = first[2]
        print('cluster_dict type:', type(cluster_dict))
        if hasattr(cluster_dict, 'keys'):
            print('cluster_dict keys count:', len(cluster_dict))
            for idx, (k,v) in enumerate(cluster_dict.items()):
                print('key', k, 'type', type(v), 'len', len(v) if hasattr(v,'__len__') else 'n/a')
                if isinstance(v, (list, tuple)) and len(v) > 1:
                    print('  item types:', [type(vv) for vv in v[:4]])
                if idx >= 3:
                    break

    if isinstance(first, (list, tuple)) and len(first) > 1 and hasattr(first[1], 'points'):
        print('first[1] has points attr; type', type(first[1]))

    if isinstance(first, dict):
        for idx, (k,v) in enumerate(first.items()):
            print('dict key', k, 'type', type(v))
            if idx >= 5:
                break

if isinstance(data, dict):
    print('dict keys sample:', list(data.keys())[:10])

print('done')
```
